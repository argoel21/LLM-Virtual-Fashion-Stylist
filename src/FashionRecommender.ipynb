```python
import warnings
warnings.filterwarnings("ignore")
import os
import torch
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from PIL import Image
from concurrent.futures import ThreadPoolExecutor
from transformers import Blip2Processor, Blip2ForConditionalGeneration, CLIPProcessor, CLIPModel
import chromadb
from chromadb.config import Settings
from sklearn.metrics.pairwise import cosine_similarity
from collections import defaultdict
import openmeteo_requests
import requests
import requests_cache
import pandas as pd
from retry_requests import retry
import openai
from openai import OpenAI
from dotenv import load_dotenv
import ast
import matplotlib.pyplot as plt
```

# Set Up Models

```python
# === Load models ===
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
blip_processor = Blip2Processor.from_pretrained("Salesforce/blip2-flan-t5-xl")
blip_model = Blip2ForConditionalGeneration.from_pretrained("Salesforce/blip2-flan-t5-xl").to(device)
```

```python
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
```

# Set Up ChromaDB

```python
# Init ChromaDB
chroma_client = chromadb.Client(Settings(anonymized_telemetry=False))

# Define sub-collections (users)
users = ["john", "gabby", "anjali", "disha"]
wardrobe_root = "../data/wardrobe_items"
collections = {}

# Setup collections per user
for user in users:
    try:
        chroma_client.delete_collection(user)
        print(f"Old collection '{user}' deleted.")
    except Exception as e:
        print(f"Warning deleting collection '{user}': {e}")
    collections[user] = chroma_client.get_or_create_collection(
        name=user,
        metadata={"hnsw:space": "cosine"}
    )
```

```python
# === BLIP-2 Metadata Questions ===
questions = {
    "description": "Give detailed description of clothing item.",
    "item": "what kind of clothing item is this? (shirt, shoes, dress, pants, shorts, skirt, jacket)",
    "formality": "Is item professional, semi-formal, or casual?",
    "occasion": "What occasions to wear this item to?",
    "season": "What season should this be worn?",
    "weather": "What kind of weather is this outfit best for? (e.g. hot, cold, rainy, sunny, mild)",
    "aesthetic": "What is the fashion aesthetic of this outfit? Choose from: indie, funky, streetwear, workout, vintage, preppy, edgy, beachy or formal.",
    "gender": "Is this outfit designed for men, women, or unisex?",
    "color": "What color is the clothing item?",
    "pattern": "What kind of pattern does this clothing item have?"
}

def get_clip_text_embedding(text):
    inputs = clip_processor(text=[text], return_tensors="pt").to(device)
    with torch.no_grad():
        vec = clip_model.get_text_features(**inputs)
        vec = vec / vec.norm(p=2, dim=-1, keepdim=True)
    return vec.cpu().numpy()[0].astype("float32")

def process_image(img_path):
    try:
        user = img_path.split(os.sep)[-2]
        collection = collections[user]

        print(f"[→] Processing: {img_path}")
        image = Image.open(img_path).convert("RGB")

        # Generate metadata using BLIP
        metadata = {"filename": img_path}
        for key, question in questions.items():
            inputs = blip_processor(images=image, text=question, return_tensors="pt").to(device)
            out = blip_model.generate(**inputs, max_new_tokens=20)
            answer = blip_processor.decode(out[0], skip_special_tokens=True).strip()
            metadata[key] = answer

        metadata = {k: str(v) for k, v in metadata.items()}

        # Embed image
        clip_inputs = clip_processor(images=image, return_tensors="pt").to(device)
        with torch.no_grad():
            img_embedding = clip_model.get_image_features(**clip_inputs)
            img_embedding = img_embedding / img_embedding.norm(p=2, dim=-1, keepdim=True)
            vector = img_embedding.cpu().numpy()[0].astype("float32")

        collection.add(
            ids=[img_path],
            embeddings=[vector],
            metadatas=[metadata]
        )

        # Embed text fields
        for key in ["description", "weather", "formality", "aesthetic", "occasion"]:
            text_val = metadata[key]
            vec = get_clip_text_embedding(text_val)
            collection.add(
                ids=[f"{img_path}:{key}"],
                embeddings=[vec],
                metadatas=[{
                    "filename": img_path,
                    "tag": key,
                    "value": text_val,
                    **metadata
                }]
            )

        print(f"[✔] Stored: {img_path}")

    except Exception as e:
        print(f"[✘] Failed: {img_path} → {e}")
        traceback.print_exc()

# === Process All Images ===
image_files = []
for user in users:
    user_dir = os.path.join(wardrobe_root, user)
    for fname in os.listdir(user_dir):
        if fname.lower().endswith((".jpg", ".jpeg", ".png")):
            image_files.append(os.path.join(user_dir, fname))

with ThreadPoolExecutor(max_workers=2) as executor:
    executor.map(process_image, image_files)
```

# AI Search

```python
def ai_outfit_search_by_type(user, user_weather, user_formality, user_aesthetic, user_occasion, top_n=5):
    if user not in collections:
        raise ValueError(f"User '{user}' does not have a ChromaDB collection initialized.")
    
    collection = collections[user]

    prompts = {
        "weather": user_weather,
        "formality": user_formality,
        "aesthetic": user_aesthetic,
        "occasion": user_occasion
    }

    weights = {
        "weather": 0.5,
        "formality": 0.2,
        "occasion": 0.2,
        "aesthetic": 0.1
    }

    query_vecs = {k: get_clip_text_embedding(v) for k, v in prompts.items()}
    scores_by_filename = {}

    for tag, query_vec in query_vecs.items():
        res = collection.query(
            query_embeddings=[query_vec],
            where={"tag": tag},
            n_results=50,
            include=["embeddings", "metadatas"]
        )

        for i in range(len(res["ids"][0])):
            meta = res["metadatas"][0][i]
            filename = meta["filename"]
            item_type = meta.get("item", "unknown").lower()
            score = cosine_similarity([query_vec], [res["embeddings"][0][i]])[0][0]
            weighted_score = weights[tag] * score

            if filename not in scores_by_filename:
                scores_by_filename[filename] = {
                    "score": weighted_score,
                    "item": item_type,
                    "metas": meta
                }
            else:
                scores_by_filename[filename]["score"] += weighted_score

    sorted_items = sorted(scores_by_filename.items(), key=lambda x: -x[1]["score"])
    grouped = defaultdict(list)

    for filename, info in sorted_items:
        grouped[info["item"]].append({
            "filename": filename,
            "score": round(info["score"], 3),
            **info["metas"]
        })

    target_types = ["shirt", "shoes", "dress", "pants", "shorts", "skirt", "jacket"]
    data = {}

    for t in target_types:
        if t in grouped:
            data[t] = grouped[t][:top_n]

    other_items = []
    for k, v in grouped.items():
        if k not in target_types:
            other_items.extend(v[:top_n])
    if other_items:
        data["other"] = other_items

    output = ''
    for item, entries in data.items():
        output += f"\n--- {item.upper()} ---\n"
        for entry in entries:
            description = entry.get("description", "").strip()
            color = entry.get("color", "").strip()
            pattern = entry.get("pattern", "").strip()
            filename = entry.get("filename", "").strip()
            output += f'Description: "{description}", Color: "{color}", Pattern: "{pattern}", Path: "{filename}"\n'

    return output
```

# Weather API

```python
def get_today_weather(city: str) -> str:
    # Get coordinates using Open-Meteo Geocoding API
    geo_resp = requests.get("https://geocoding-api.open-meteo.com/v1/search", params={"name": city})
    geo_data = geo_resp.json()

    if "results" not in geo_data or len(geo_data["results"]) == 0:
        return f"[✘] City not found: {city}"

    lat = geo_data["results"][0]["latitude"]
    lon = geo_data["results"][0]["longitude"]
    location_name = geo_data["results"][0]["name"]

    # Setup Open-Meteo weather API with caching and retry
    cache_session = requests_cache.CachedSession('.cache', expire_after=3600)
    retry_session = retry(cache_session, retries=5, backoff_factor=0.2)
    openmeteo = openmeteo_requests.Client(session=retry_session)

    # Weather API params
    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": lat,
        "longitude": lon,
        "daily": [
            "temperature_2m_max",
            "temperature_2m_min",
            "uv_index_max",
            "precipitation_sum",
            "precipitation_probability_max",
            "wind_speed_10m_max"
        ],
        "timezone": "auto"
    }

    # Call weather API
    responses = openmeteo.weather_api(url, params=params)
    response = responses[0]

    daily = response.Daily()
    tmax = daily.Variables(0).ValuesAsNumpy()
    tmin = daily.Variables(1).ValuesAsNumpy()
    uv = daily.Variables(2).ValuesAsNumpy()
    precip = daily.Variables(3).ValuesAsNumpy()
    precip_prob = daily.Variables(4).ValuesAsNumpy()
    wind = daily.Variables(5).ValuesAsNumpy()

    dates = pd.date_range(
        start=pd.to_datetime(daily.Time(), unit="s", utc=True),
        end=pd.to_datetime(daily.TimeEnd(), unit="s", utc=True),
        freq=pd.Timedelta(seconds=daily.Interval()),
        inclusive="left"
    )

    df = pd.DataFrame({
        "date": dates,
        "temperature_2m_max": ((tmax * 9 / 5) + 32).round().astype(int),
        "temperature_2m_min": ((tmin * 9 / 5) + 32).round().astype(int),
        "uv_index_max": uv.round().astype(int),
        "precipitation_sum": precip.round().astype(int),
        "precipitation_probability_max": precip_prob.round().astype(int),
        "wind_speed_10m_max": wind.round().astype(int)
    })

    today = pd.Timestamp.now(tz="UTC").normalize()
    today_weather = df[df["date"] == today]
    
    if today_weather.empty:
        today_weather = df.iloc[[0]]

    row = today_weather.iloc[0]

    return (
        f"Weather for {location_name} today:\n"
        f"High: {row.temperature_2m_max}°F | Low: {row.temperature_2m_min}°F\n"
        f"UV Index: {row.uv_index_max}\n"
        f"Precipitation: {row.precipitation_sum} mm | Chance: {row.precipitation_probability_max}%\n"
        f"Wind Speed Max: {row.wind_speed_10m_max} km/h"
    )
```

# Chat Interaction

```python
# Load .env file
load_dotenv()

# Initialize the client
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))
```

```python
df = pd.read_csv(
    "../data/pinterest_trends.csv",
    skiprows=10,
    engine="python",
    encoding="utf-8",
    on_bad_lines='warn'
)

# Optional: add trend context from the Pinterest dataset
trend_examples = df["Trend"].dropna().sample(5).tolist()
trend_context = ", ".join(trend_examples)
```

```python
def recommend_outfit(user_query, user, aesthetic='none', location='Charlottesville'): 
    weather_context = get_today_weather(location)

    weather_message = f"Describe the weather very briefly from this context: {weather_context}. Examples: cold and rainy, warm and sunny, etc."
    weather_response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": weather_message}],
        temperature=0.15
    )
    weather_search = weather_response.choices[0].message.content.strip()

    formality_message = f"From the user's query, return if the event is casual, semi-professional, or professional: {user_query}"
    formality_response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": formality_message}],
        temperature=0.01
    )
    formality_search = formality_response.choices[0].message.content.strip()

    occasion_message = f"From the user's query, return what the occasion is, very brief: {user_query}"
    occasion_response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": occasion_message}],
        temperature=0.15
    )
    occasion_search = occasion_response.choices[0].message.content.strip()

    wardrobe_context = ai_outfit_search_by_type(
        user=user,
        user_weather=weather_search,
        user_formality=formality_search,
        user_aesthetic=aesthetic,
        user_occasion=occasion_search,
        top_n=3
    )

    outfit_prompt = (
        f"Given this user query: '{user_query}', "
        f"and the following wardrobe items:\n\n{wardrobe_context}\n\n"
        f"Suggest the best outfit from these items based on the event, weather: ('{weather_context}'), "
        f"formality ('{formality_search}'), and aesthetic ('{aesthetic}'). "
        f"Be concise and describe why each item was chosen, especially for weather. Do not recommend anything not needed."
        f"Also consider what is trending right now: {trend_context}"
    )

    outfit_response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": outfit_prompt}],
        temperature=0.7
    )
    outfit_rec = outfit_response.choices[0].message.content.strip()
    
    files_prompt = (
        f"Given the following response, return the file paths to the items recommended. "
        f"Here is the response: {outfit_rec}"
        f"Here is the wardrobe context: {wardrobe_context}"
        f"Return them in a list and return no other text, like '['path1','path2']'"
    )

    files_response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": files_prompt}],
        temperature=0.25
    )
    files = files_response.choices[0].message.content
    
    image_paths = ast.literal_eval(files)
    
    for path in image_paths:
        img = Image.open(path)
        plt.imshow(img)
        plt.title(path)
        plt.axis('off')
        plt.show()
    
    print(outfit_rec)
```

```python
user_query_1 = "I'm going for a long walk through the park."

recommend_outfit(
    user_query=user_query_1, 
    user='john', 
    aesthetic='vintage', 
    location='Charlottesville'
)
```
